## Email Classification
Develope a tool which will classify an email.

### Tasks:
* Create a function classify_emails that takes a list of emails as an argument.
* Inside the function, categorize emails into "Important," "Promotions," and "Spam" based on keywords and content analysis.
* Loop through the emails and identify keywords to determine their category.
* If an email contains keywords indicating importance, assign it as "Important."
* If an email contains keywords related to promotions or deals, assign it as "Promotions."
* If an email is suspected to be spam based on certain keywords or patterns, assign it as "Spam."
* Return three lists of emails for each category.
* Print the lists.

```python
def classify_emails(emails):
    # Spam Keywords
    spam=[]
    imp=[]
    promo=[]
    # Spam Keywords
    spam_keywords = [
        "nigerian prince",
        "free money",
        "enlarge",
        "make money fast",
        "guaranteed",
        "click here",
        "limited time offer",
        "congratulations, you've won",
        "unsubscribe"
    ]

    # Promotions Keywords
    promotions_keywords = [
        "sale",
        "discount",
        "deal",
        "offer",
        "promo",
        "clearance",
        "coupon",
        "limited time",
        "special",
        "exclusive"
    ]

    # Important Keywords
    important_keywords = [
        "urgent",
        "action required",
        "deadline",
        "confirmation",
        "invoice",
        "payment",
        "meeting",
        "appointment",
        "contract",
        "emergency"
    ]

    for i in emails:
        for f in spam_keywords:
            if i.lower().__contains__(f):
                spam.append(i)
        for f in important_keywords:
            if i.lower().__contains__(f):
                imp.append(i)
        for f in promotions_keywords:
            if i.lower().__contains__(f):
                promo.append(i)

    return spam, imp, promo

test_emails = [
    "Hello, urgent meeting tomorrow.",
    "Get a special discount on our products!",
    "Congratulations! You've won $1000!",
    "Enlarge your savings with our limited time offer.",
    "Invoice for your recent purchase.",
    "Nigerian prince needs your help.",
    "Reminder: Action required for your contract.",
    "Exclusive deals just for you.",
    "Free money offer inside.",
    "Important deadline approaching."
]

spam, important, promotions = classify_emails(test_emails)

print("Important Emails:")
print(important)
print("\nPromotions Emails:")
print(promotions)
print("\nSpam Emails:")
print(spam)
```

#Exercise
Use the dataset of Game, we have explored earlier and perform following operations.

* Calculate the total global sales for all games in the dataset.
* Find the game with the highest global sales.
* Determine the average sales for each region (NA_Sales, EU_Sales, JP_Sales, Other_Sales).
* Find the top 5 most popular game genres based on global sales.
* Calculate the total sales for each genre in each region and find the most popular genre in North America (NA_Sales).
* Display last 10 rows.
* Determine the top 3 gaming platforms with the highest global sales.
* Calculate the market share (percentage of global sales) for each genre.
* Calculate the correlation coefficient between NA_Sales and EU_Sales. (.corr function)

```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/TirendazAcademy/PANDAS-TUTORIAL/main/DataSets/vgsalesGlobale.csv"
df = pd.read_csv(url)

# Display the DataFrame
df
```

**TOTAL GLOBAL SALES**

```python
df["Global_Sales"].sum()
```

```python
ds=df[df["Global_Sales"]==df["Global_Sales"].max()]
ds
```

Determine the average sales for each region (NA_Sales, EU_Sales, JP_Sales, Other_Sales).

```python
print("\nNA SALES\n",df["NA_Sales"].mean(),"\nEU SALES\n",df["EU_Sales"].mean(),"\nJP SALES\n",df["JP_Sales"].mean(),"\nOther SALES\n",df["Other_Sales"].mean())
```

Find the top 5 most popular game genres based on global sales.

```python
a=df.sort_values(by="Global_Sales",ascending=False)
print(a["Genre"].head())
```

Calculate the total sales for each genre in each region and find the most popular genre in North America (NA_Sales).





```python
genre_sales = df.groupby('Genre').agg({
    'NA_Sales': 'sum','EU_Sales': 'sum','JP_Sales': 'sum','Other_Sales': 'sum','Global_Sales': 'sum'
})
print(genre_sales)

print("\n\n\n\nMost popular genre",genre_sales['NA_Sales'].idxmax())
```

last 10 rows

```python
df.tail(10)
```

Determine the top 3 gaming platforms with the highest global sales.

```python
platform_sales = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)
print("Top 3 platforms ",platform_sales.head(3))
```

**Calculate the market share (percentage of global sales) for each genre.**

```python
suum=df["Global_Sales"].sum()
share=(df.groupby("Genre")["Global_Sales"].sum()/suum)*100
share
```

***Calculate the correlation coefficient between NA_Sales and EU_Sales.***

```python
print([df["NA_Sales"].corr(df["EU_Sales"])])
```
